```python
# The count of each gender in the gym 
```

```python
sns.countplot (data=df, x='Gender', color='r')
plt.title('count of Gender')
plt.show()
```

```python
sns.histplot(data=df, x='Age', kde=True,palette='inferno')
plt.title("Distrbution Of Age")
```

```python
#this chart shows the age Distribution of the gym both grinders at age between  40 and 50 Are healthy concise
```

```python
#Relation between weight and BMI
```

```python
sns.lmplot (data= df, x='Weight (kg)' , y='BMI')
plt.title ('Relation between weight and BMI')
plt.show()
```

```python
#The LM plot reveals a weak positive correlation between age and BMI, indicating that BMI slightly increases with age.
#However, the relationship appears to be modest, with considerable variation across the data
```

```python
sns.barplot(data= df, x='Workout_Type',y='Age',palette='Set3')
plt.title ('Age and workout type')

plt.show()

```

```python
#in this bar chart shows the age and workout type most of them are tying to hiit , yoga and stength tinging
```

```python
sns.lineplot(x=df['Workout_Type'],y=df['Calories_Burned'])
plt.xticks(rotation=90)
plt.title('Relationship Between calories Burned and Workout Type')
plt.show()
```

```python
#in this line chart showes the hiit workout effect more caloriburneds 
```

```python
corr=df.corr(numeric_only=True)
corr


```

```python
sns.heatmap(corr)
plt.show()
```

```python
sns.scatterplot (data=df, x='Age',y='Fat_Percentage',color='red')
plt.xlabel('Age')
plt.ylabel('Fat Percentage')
plt.title('Age vs Fat Percentage')
plt.show()
```

```python
#The scatter plot shows a weak positive correlation between age and body fat percentage, with a slight increase in fat percentage as age rises. However,
#the data points exhibit significant variability, indicating other factors may influence fat percentage.
```

```python
sns.violinplot (data = df , x='Gender',y='Water_Intake (liters)' ,color='yellow')
plt.xlabel('Gender')
plt.ylabel('Water_Intake (liters)')
plt.title('Water intake Distribution by Gender')

plt.show()
```

```python
#in this violinplot shows that males have a higher average water intake compared to females. 
#Males consistently consume more water, with a noticeable difference in the quantities consumed.
```
